```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
tennis_2019 = pd.read_csv('tennis_atp/atp_matches_2019.csv', delimiter=',')
tennis_2018 = pd.read_csv('tennis_atp/atp_matches_2018.csv', delimiter=',')
tennis_2017 = pd.read_csv('tennis_atp/atp_matches_2017.csv', delimiter=',')
tennis_2016 = pd.read_csv('tennis_atp/atp_matches_2016.csv', delimiter=',')
tennis_2015 = pd.read_csv('tennis_atp/atp_matches_2015.csv', delimiter=',')
```

# Court Surfaces

```python
hard_court_total = final_tennis[final_tennis['surface'] == 'Hard'] 
clay_court_total = final_tennis[final_tennis['surface'] == 'Clay']
grass_court_total = final_tennis[final_tennis['surface'] == 'Grass']
```

```python
# Lets check out which how the countries as a whole perform on the different types of surfaces
# SUI
sui_hard = final_tennis[((final_tennis['surface'] == 'Hard') & (final_tennis['winner_ioc'] == 'SUI'))]
sui_clay = final_tennis[((final_tennis['surface'] == 'Clay') & (final_tennis['winner_ioc'] == 'SUI'))] 
sui_grass = final_tennis[((final_tennis['surface'] == 'Grass') & (final_tennis['winner_ioc'] == 'SUI'))] 

sui_hard_percent = len(sui_hard) / len(hard_court_total) * 100
sui_clay_percent = len(sui_clay) / len(clay_court_total) * 100
sui_grass_percent = len(sui_grass) / len(grass_court_total) * 100

print(f"Players from SUI won {sui_hard_percent:.2f}% of tournaments on hard surfaces from 2015 - 2019.")
print(f"Players from SUI won {sui_clay_percent:.2f}% of tournaments on clay surfaces from 2015 - 2019.")
print(f"Players from SUI won {sui_grass_percent:.2f}% of tournaments on grass surfaces from 2015 - 2019.")
```

```python
# ESP
esp_hard = final_tennis[((final_tennis['surface'] == 'Hard') & (final_tennis['winner_ioc'] == 'ESP'))] 
esp_clay = final_tennis[((final_tennis['surface'] == 'Clay') & (final_tennis['winner_ioc'] == 'ESP'))] 
esp_grass = final_tennis[((final_tennis['surface'] == 'Grass') & (final_tennis['winner_ioc'] == 'ESP'))] 

esp_hard_percent = len(esp_hard) / len(hard_court_total) * 100
esp_clay_percent = len(esp_clay) / len(clay_court_total) * 100
esp_grass_percent = len(esp_grass) / len(grass_court_total) * 100

print(f"Players from ESP won {esp_hard_percent:.2f}% of tournaments on hard surfaces from 2015 - 2019.")
print(f"Players from ESP won {esp_clay_percent:.2f}% of tournaments on clay surfaces from 2015 - 2019.")
print(f"Players from ESP won {esp_grass_percent:.2f}% of tournaments on grass surfaces from 2015 - 2019.")
```

```python
# SRB
srb_hard = final_tennis[((final_tennis['surface'] == 'Hard') & (final_tennis['winner_ioc'] == 'SRB'))] 
srb_clay = final_tennis[((final_tennis['surface'] == 'Clay') & (final_tennis['winner_ioc'] == 'SRB'))]
srb_grass = final_tennis[((final_tennis['surface'] == 'Grass') & (final_tennis['winner_ioc'] == 'SRB'))] 

srb_hard_percent = len(srb_hard) / len(hard_court_total) * 100
srb_clay_percent = len(srb_clay) / len(clay_court_total) * 100
srb_grass_percent = len(srb_grass) / len(grass_court_total) * 100

print(f"Players from SRB won {srb_hard_percent:.2f}% of tournaments on hard surfaces from 2015 - 2019.")
print(f"Players from SRB won {srb_clay_percent:.2f}% of tournaments on clay surfaces from 2015 - 2019.")
print(f"Players from SRB won {srb_grass_percent:.2f}% of tournaments on grass surfaces from 2015 - 2019.")
```
