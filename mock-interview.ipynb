```python
vec1 = np.array([1,4,6,8])
vec2 = np.array([1,6,3,9])
string = 'l2'

def dist_btwn_vectors(vec1, vec2, string):
    if string == 'l1': # manhatten
        return abs(vec1 - vec2).sum()
    else: 
        return np.sqrt(((vec1 - vec2)**2).sum())
    
print(dist_btwn_vectors(vec1, vec2, string))
```

```python
n = 9
def is_prime(n):
    if n == 2:
        return True
    
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
    
print(is_prime(n))
```
