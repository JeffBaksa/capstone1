```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
tennis_2019 = pd.read_csv('tennis_atp/atp_matches_2019.csv', delimiter=',')
copy_tennis_2019 = tennis_2019.copy()
```

```python
fig = plt.figure(figsize=(20, 20))
ax = fig.gca()
tennis_2019.hist(ax = ax)
```

```python
# Number of individual players who played in at least one match in 2019
total_players = tennis_2019[['winner_name', 'winner_ioc']] # 241 unique players

#pd.unique(total_players.groupby(['winner_ioc', 'winner_name']))
country_totals = total_players.groupby('winner_ioc')['winner_name'].nunique().sort_values()
country_totals
```

```python
# Replacing the NaNs with 0s for columns that will be used later
tennis_2019['w_ace'] = tennis_2019['w_ace'].fillna(0)
tennis_2019['l_ace'] = tennis_2019['l_ace'].fillna(0)

tennis_2019['w_bpSaved'] = tennis_2019['w_bpSaved'].fillna(0)
tennis_2019['l_bpSaved'] = tennis_2019['l_bpSaved'].fillna(0)

tennis_2019['w_1stIn'] = tennis_2019['w_1stIn'].fillna(0)
tennis_2019['l_1stIn'] = tennis_2019['l_1stIn'].fillna(0)
```

## Surfaces

```python
# Want to count the number of wins on each surface...then identify where the winner originates from (winner_ioc)
pd.unique(tennis_2019['surface']) # 3 different surfaces
```

```python
pd.unique(tennis_2019['winner_ioc'])
```

```python
# Checking the popular regions with each corresponding surface
# Count the number of winners from 'AUS' and 'USA' where tournament surface == 'Hard'
hard_court_usa_aus = tennis_2019[((tennis_2019['surface'] == 'Hard') & (tennis_2019['winner_ioc'] == 'AUS') | (tennis_2019['winner_ioc'] == 'USA'))] # 311
hard_court_total = tennis_2019[tennis_2019['surface'] == 'Hard'] # 1626

hard_court_region_percent = len(hard_court_usa_aus) / len(hard_court_total) * 100
print(f"Players from USA or AUS won {hard_court_region_percent:.2f}% of tournaments on hard court surfaces in 2019.")
```

```python
# Count the number of winners from 'ESP' and 'ARG' where tournament surface == 'Clay'
clay_court_esp_arg = tennis_2019[((tennis_2019['surface'] == 'Clay') & (tennis_2019['winner_ioc'] == 'ESP') | (tennis_2019['winner_ioc'] == 'ARG'))] # 260
clay_court_total = tennis_2019[tennis_2019['surface'] == 'Clay'] # 828

clay_court_region_percent = len(clay_court_esp_arg) / len(clay_court_total) * 100
print(f"Players from ESP or ARG won {clay_court_region_percent:.2f}% of tournaments on clay surfaces in 2019.")
```

```python
# Count the number of winners from 'GBR' and 'USA' where tournament surface == 'Grass'
grass_court_gbr_usa = tennis_2019[((tennis_2019['surface'] == 'Grass') & (tennis_2019['winner_ioc'] == 'GBR') | (tennis_2019['winner_ioc'] == 'USA'))] # 211
grass_court_total = tennis_2019[tennis_2019['surface'] == 'Grass'] # 327

grass_court_region_percent = len(grass_court_gbr_usa) / len(grass_court_total) * 100
print(f"Players from GBR or USA won {grass_court_region_percent:.2f}% of tournaments on grass surfaces in 2019.")
```

```python
# Avg aces in victories
fig, ax = plt.subplots(figsize=(10, 4)) 

ax.bar('Federer', fed['w_ace'].mean())
ax.bar('Nadal', nadal['w_ace'].mean())
ax.bar('Djokovic', djok['w_ace'].mean())

ax.set_title('Avg Aces per Matches Won (2019)', fontsize=14)
ax.set_xlabel('Player')
ax.set_ylabel('Num of Aces')
```

```python
# Avg aces in losses
fig, ax = plt.subplots(figsize=(12, 4)) 

ax.bar('Federer', copy_fed['l_ace'].mean())
ax.bar('Nadal', copy_nadal['l_ace'].mean())
ax.bar('Djokovic', copy_djok['l_ace'].mean())

ax.set_title('Avg Aces per Matches Lost (2019)', fontsize=14)
ax.set_xlabel('Player')
ax.set_ylabel('Num of Aces')
```

```python
# Compare the distribution of aces in wins for each player

fig, federer_ax = plt.subplots(figsize=(12, 4)) 
federer_ax.hist(fed['w_ace'])
federer_ax.set_title('Federer Frequency of Aces per Matches Won (2019)', fontsize=14)
federer_ax.set_xlabel('Num of Aces')
federer_ax.set_ylabel('Frequency')

fig, nadal_ax = plt.subplots(figsize=(12, 4)) 
nadal_ax.hist(nadal['w_ace'])
nadal_ax.set_title('Nadal Frequency of Aces per Matches Won (2019)', fontsize=14)
nadal_ax.set_xlabel('Num of Aces')
nadal_ax.set_ylabel('Frequency')

fig, djok_ax = plt.subplots(figsize=(12, 4)) 
djok_ax.hist(djok['w_ace'])
djok_ax.set_title('Djokovic Frequency of Aces per Matches Won (2019)', fontsize=14)
djok_ax.set_xlabel('Num of Aces')
djok_ax.set_ylabel('Frequency')
```

```python
# Compare avgs of 1st serves in for each of the three players

fig, ax = plt.subplots(figsize=(12, 4)) 

ax.bar('Federer', fed['w_1stIn'].mean())
ax.bar('Nadal', nadal['w_1stIn'].mean())
ax.bar('Djokovic', djok['w_1stIn'].mean())

ax.set_title('Avg 1st Serves In per Matches Won (2019)', fontsize=14)
ax.set_xlabel('Player')
ax.set_ylabel('Num of 1st Serves In')
```

```python
# Compare avgs of bp saved in for each of the three players

fig, ax = plt.subplots(figsize=(12, 4)) 

ax.bar('Federer', fed['w_bpSaved'].mean())
ax.bar('Nadal', nadal['w_bpSaved'].mean())
ax.bar('Djokovic', djok['w_bpSaved'].mean())

ax.set_title('Avg Break Points Saved per Matches Won (2019)', fontsize=14)
ax.set_xlabel('Player')
ax.set_ylabel('Num of Break Points Saved')
```

```python
# Grouped barchart of wins/losses for each player
fig, ax = plt.subplots(figsize=(12, 4))

x = ['Federer', 'Nadal', 'Djok']
wins = [fed_win_num, nadal_win_num, djok_win_num] #y1
losses = [fed_loss_num, nadal_loss_num, djok_loss_num] #y2
width = 0.5

p1 = ax.barh(x, wins, width, color='tab:orange', label='Wins')
p2 = ax.barh(x, losses, width, color='tab:blue', label='Losses')
ax.set_title('Wins and Losses', fontsize=14)
ax.set_xlabel('Num of Matches Played')
ax.set_ylabel('Player')
ax.legend()

ax.bar_label(p1, label_type='center')
ax.bar_label(p2, label_type='center')
```
